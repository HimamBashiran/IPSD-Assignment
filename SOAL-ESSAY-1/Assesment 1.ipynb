Keterangan :
Jika inputan berupa tanggal lahir maka anda harus menghitung usia anda saat
assessment ini dilakukan.
Untuk menghitung luas lingkaran anda dapat menggunakan rumus luas lingkaran
yaitu 3,14 x jari-jari x jari-jari 

```python
import datetime

def hitung_usia(tanggal_lahir):
    hari_ini = datetime.date.today()
    usia = hari_ini.year - tanggal_lahir.year
    if hari_ini.month < tanggal_lahir.month or (hari_ini.month == tanggal_lahir.month and hari_ini.day < tanggal_lahir.day):
        usia -= 1
    return usia

def hitung_luas_lingkaran(jari_jari):
    return 3.14 * jari_jari * jari_jari

# Input data
nama = input("Masukkan nama: ")
nim = input("Masukkan NIM: ")

pilihan = input("Pilih input \n1: Usia \n2: Tanggal Lahir ")
if pilihan == "1":
    usia = int(input("Masukkan usia: "))
elif pilihan == "2":
    tanggal_lahir = input("Masukkan tanggal lahir (format: DD-MM-YYYY): ")
    tanggal_lahir = datetime.datetime.strptime(tanggal_lahir, "%d-%m-%Y").date()
    usia = hitung_usia(tanggal_lahir)
else:
    print("Pilihan tidak valid")
    exit()

angka_favorit = float(input("Masukkan angka favorit: "))

# Hitung luas lingkaran
luas_lingkaran = hitung_luas_lingkaran(angka_favorit)

# Output
print("\nOUTPUT")
print(f"Nama: {nama}")
print(f"NIM: {nim}")
print(f"Usia: {usia}")
print(f"Luas lingkaran berdasarkan angka favorit: {luas_lingkaran:.2f}")
```

Lakukan eksplorasi data pada data genangan air Surabaya.csv dengan
membuat program-program python dan interpretasi dari output program
yang anda buat. Sehingga anda dapat menjawab pertanyaan-pertanyaan
berikut ini :

a. Tentukan berapa nilai maksimum, minimum dan rata-rata dari luas
genangan, tinggi genangan dan lama genangan air. ( 5)

b. Buatlah pie chart untuk membandingkan banyaknya peristiwa genangan
berdasarkan rayonnya. (5)

c. Bagaimana hubungan antara luas genangan dan skor genangan, dan apakah
ada hubungan antara luas genangan air dan lama genangan air? (5)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca file CSV
df = pd.read_csv('data genangan air Surabaya.csv')

# Soal 2a: Nilai maksimum, minimum, dan rata-rata dari luas genangan, tinggi genangan dan lama genangan air
metrics = ['Luas genangan', 'Tinggi genangan', 'Lama genangan']
stats = df[metrics].agg(['max', 'min', 'mean'])
print("2a. Statistik genangan:")
print(stats)
```

```python
# Soal 2b: Buatlah pie chart untuk membandingkan banyaknya peristiwa genangan berdasarkan rayonnya
rayon_counts = df['Rayon'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(rayon_counts.values, labels=rayon_counts.index, autopct='%1.1f%%')
plt.title('Persentase Peristiwa Genangan berdasarkan Rayon')
plt.axis('equal')
plt.show()
```

```python
# Soal 2c: Hubungan antara luas genangan, skor genangan, dan lama genangan
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(df['Luas genangan'], df['Skor Luas Genangan'])
plt.xlabel('Luas Genangan')
plt.ylabel('Skor Luas Genangan')
plt.title('Hubungan Luas Genangan dan Skor Genangan')

plt.subplot(122)
plt.scatter(df['Luas genangan'], df['Lama genangan'])
plt.xlabel('Luas Genangan')
plt.ylabel('Lama Genangan')
plt.title('Hubungan Luas Genangan dan Lama Genangan')
plt.show()

#correlation
# Menghitung matriks korelasi
correlation_matrix = df[['Luas genangan', 'Tinggi genangan', 'Lama genangan', 
                          'Skor Luas Genangan', 'Skor Tinggi Genangan', 
                          'Skor Lama Genangan']].corr()

# Mengatur ukuran figure
plt.figure(figsize=(10, 8))

# Membuat heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

# Menambahkan judul
plt.title('Heatmap Korelasi Variabel Genangan Air di Surabaya')
plt.show()

print(correlation_matrix)
```

Interpretasi: dari korelasi diatas, dapat kita lihat bersama bahwa untuk variabel tinggi genangan dengan lama genangan berkorelasi cukup kuat yang artinya jika tinggi genangan besar, maka lama air itu tergenang juga besar.

```python
# Soal 3: Praproses data
# Baca file CSV dengan separator ';'
df_modified = pd.read_csv('data genangan air Surabaya.csv')

# Soal 3a: Apakah ada duplikasi data? Jika ya/tidak apa yang anda lakukan?
duplicates = df_modified.duplicated().sum()
print(f"Jumlah data duplikat: {duplicates}")
```

Interpretasi: Tidak ada duplikasi data, artinya data sudah bersih dan tidak perlu melakukan apapun untuk menghindari rusaknya data

```python
# Soal 3b: Apakah ada missing value atau noise atau nilai bias pada data tersebut? Jika ya/tidak apa yang anda lakukan?
df_modified.isna().sum()
```

Interpretasi: Tidak ada missing values atau data yang hilang, artinya kita tidak perlu melakukan drop ataupun imputasi

```python
# Membuat dictionary mapping untuk Kecamatan, Kelurahan, dan Rayon
kecamatan_map = {}
kelurahan_map = {}
rayon_map = {}

def create_mapping(column):
    unique_values = sorted(set(column))
    
    return {v: i + 1 for i, v in enumerate(unique_values)}

kecamatan_map = create_mapping(df_modified['Kecamatan'])
kelurahan_map = create_mapping(df_modified['Kelurahan'])
rayon_map   = create_mapping(df_modified['Rayon'])

# Melakukan pemetaan nilai ke index numerik
df_modified['Kecamatan'] = df_modified['Kecamatan'].apply(lambda x : kecamatan_map[x])
df_modified['Kelurahan'] = df_modified['Kelurahan'].apply(lambda x : kelurahan_map[x])
df_modified['Rayon']     = df_modified['Rayon'].apply(lambda x : rayon_map[x])

print("Setelah proses mapping:")
print(df_modified.info())
print(df_modified.head())

# Verifikasi hasil mapping
assert all(isinstance(x, int) and isinstance(y, int) 
           for x,y in zip(df_modified['Kecamatan'], range(len(kecamatan_map))))
assert all(isinstance(x, int) and isinstance(y, int) 
           for x,y in zip(df_modified['Kelurahan'], range(len(kelurahan_map))))
assert all(isinstance(x, int) and isinstance(y, int) 
           for x,y in zip(df_modified['Rayon'], range(len(rayon_map))))
```

```python
#Boxplot
num_features = len(df_modified.columns)

num_cols = 4  
num_rows = (num_features // num_cols) + (num_features % num_cols > 0)

plt.figure(figsize=(num_cols * 5, num_rows * 4)) 

for i, feature in enumerate(df_modified.columns):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.boxplot(y=df_modified[feature], color='red')
    plt.title(f'Boxplot of {feature}', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Winsorize to handling outliers
columns_to_plot = [col for col in df_modified.columns if col != 'jumlah_promosi']
num_cols = len(columns_to_plot)
num_rows = (num_cols + 3) // 4

fig_width = 12
fig_height = 3 * num_rows

fig, axes = plt.subplots(num_rows, 4, figsize=(fig_width, fig_height))

for i, column in enumerate(columns_to_plot):
    ax = axes[i // 4, i % 4] 
    ax.hist(df_modified[column], bins=20, color='skyblue', alpha=0.7, edgecolor='black', label='df_modified')
    ax.set_title(f'Histogram {column}', fontsize=10)
    ax.set_xlabel('Value', fontsize=8)
    ax.set_ylabel('Frequency', fontsize=8)
    ax.legend()
    ax.grid(True)

for i in range(num_cols, num_rows * 4):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout(pad=2.0)
plt.show()
```

Interpretasi: setelah mengubah tipe data Kecamatan, Kelurahan, dan Rayon. Kita bisa mengecek outliers menggunakan boxplot dan melakukan handling outliers menggunakan winsorize
